# correlation is either positive , Negative and  last one is nocorrelation
# positive=> increase of one variable increase the another varible and decrease huda arko ni descrese 
# Negative=> yauta increase huda arko descrese hunxa rw (and vice versa)
# No correaltion=> there is no relationship(zigzag pattern dekhinxa )

compute=(-1,1)
close to -1:negative correlation
close to 1: positive correaltion
0: no correlation

```python
# for the computation of the above plot we use the function (.corr) 
corage_Salary=df['Age'].corr(df['Salary'])# yauta ma yauta parameter arko corr ma arko parameter 
print(corage_Salary)
```

0-0.99:weak correaltion
0.5-0.699:Average_correlation
0.7: strong correlation 

same values for the negative side (weak, Average, strong)
