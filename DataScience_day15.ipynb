# Data Scaling 
data ko range yautai xaina data ko range xuttai xuttai xa teo bela scaling garinxa assume the column 

Age     Insulin
21        0
80        450

assume there is someone wich has age
(40,42) and insulin (300,100)
here the age has 40 42 they are very close  but at the same time the colum insulin are very far 
to make them reside in the same range we need scaling 


there are some techniques
1. min max scaling
=>( x-xmin)/(xmax-xmin)   in the range of (0,1)

a +[ ((x - x_min) / (x_max - x_min)) ]* (b - a) 

(a,b): desired range (e.g., 0 to 1)

3. standardization (Z-score normalization)
   =>(x-mew)/sd
   
Note: scaling is done after train test split 
standard scaler ley mean =0, sd=1
