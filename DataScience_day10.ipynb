```python
from sklearn.linear_model import LinearRegression
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error,r2_score
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("Salary_Data.csv")
data.head()
```

```python
data.isnull().sum()                                         # to check if the nulll value exist or not 
```

```python
                                        # to check if the nulll value exist or not 
data.dropna(inplace=True)
```

```python
data.isnull().sum()     
```

```python
x=data[['Age']]
y=data.Salary
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=7)
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```

```python
ypred=model.predict(xtrain)
ypred
```

```python
plt.plot                     # syntax for the line plot 
```

```python
plt.scatter(data.Age,data.Salary)
plt.plot(xtrain,ypred,color='red')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age Vs salary')
```

```python
yprediction=model.predict(xtrain)
```

```python
plt.scatter(data.Age,data.Salary)              # this is the scatter plot of the whole data 

plt.plot(xtrain,yprediction,color='red')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age Vs salary')
```
