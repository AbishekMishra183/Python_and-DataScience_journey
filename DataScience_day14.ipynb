```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score

from sklearn.neighbors import KNeighborsClassifier
```

```python
data=sns.load_dataset('iris')
```

```python
data.head()
```

```python
X=data.drop('species',axis=1)   # species bhayak aru sabbi lai x ma rakhne ko lagi drop garyako
y=data.species
```

```python
xtrain,xtest,ytrain, ytest= train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
knn=KNeighborsClassifier(n_neighbors=9) # n_neighbors vanyako k ko value dino ho in the knn
# default k ko value chai 5 linxa
knn.fit(xtrain,ytrain)
```

```python
help(KNeighborsClassifier)
```

```python
ypred=knn.predict(xtest)
```

```python
ypred
```

```python
cm=confusion_matrix(ytest,ypred)
```

```python
cm
```

```python
accuracy=accuracy_score(ytest,ypred)
print(f'Accuracy:{accuracy*100}')
```
