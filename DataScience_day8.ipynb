```python
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("Salary_Data.csv")
data
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
x=data[['Age']]
y=data.Salary
```

```python
# split to train and train
# random_state= same split rahairahanxa matlab harayak time train rw test value split huda fix hudaina yesle chai fix garna help garxa
#7 matrw rakhno parxa vannae hudaina  j rakhna ni payainxa tera jaba samma 7 lai change garidana teti bela samma train test value change hudaian
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=7)
```

```python
xtrain
```

```python
ytrain
```

```python
model=LinearRegression()
model.fit(xtrain,ytrain)
```

```python
# aba evaluation garna ko lagi hami xtest rw ytest dinxau for the predection(ypred)
```

```python
ypredection=model.predict(xtest)
```

```python
ypredection
```

```python
xtest      ytest  ypredection
```

```python
mae = mean_absolute_error(ytest, ypredection)
print(f"MAE:{mae:.2f}")
```
