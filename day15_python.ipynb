```python
# get all the prices below 5 in the USD
us_price={'milk':2.05,'bread':2.6,'butter':3.6,'mobile':50,'television':1000,'refrigenator':700}

# non pythonic way
price_below_5={}
for product, price in us_price.items():
    if price<5:
        price_below_5.update({product:price})
print (price_below_5)


```

```python
# pythonic way
us_price={'milk':2.05,'bread':2.6,'butter':3.6,'mobile':50,'television':1000,'refrigenator':700}
price_below_5={
    product:price
    for product, price in us_price.items()
    if price<5

}# dictionary comphernsion
print (price_below_5)
```
