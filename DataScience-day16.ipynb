```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

3.test_data(transform)

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.8,random_state=22)
```

```python
xtrain.head()
```

```python
scaler=StandardScaler()
xtrain_s=scaler.fit_transform(xtrain)
xtest_s=scaler.transform(xtest)
```

```python
xtrain_s
```

```python
xtrains=pd.DataFrame(xtrain_s,columns=xtrain.columns) # to see the output in the dataframe format
xtrains.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(xtrain_s,ytrain)
```

```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
data=sns.load_dataset('iris')
```

```python
data.head()
```

```python
x=data.drop('species',axis=1)
y=data.species
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
dmodel=DecisionTreeClassifier()
dmodel.fit(xtrain,ytrain)
```

```python
dmodel.predict([[4,3.1,1,0.5]])
```

```python
ypredection=dmodel.predict(xtest)
```

```python
ypredection
```

```python
accuracy=accuracy_score(ytest,ypredection)
print(f'Accuracy:{accuracy:.2%}')
```

```python
from sklearn.tree import plot_tree   # to make the decision tree 
import matplotlib.pyplot as plt 
```

```python
clas =list(y.unique())
clas
#clas=data['species'].unique
```

```python
feature=list(x.columns)
feature
```

```python
plot_tree(dmodel,feature_names=feature,class_names=clas,filled=True)    # demodel is the name of the variable 
plt.show()
```
